# **FUZZY SETS**

```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

#### **Sets for ToF sensor range** $[0, \dots, 10]$ (meters)

```python
# Definir el universo para la variable difusa ToF_sensor_range
universe_ToF_sensor_range = np.arange(0, 10, 0.01)
ToF_sensor_range = ctrl.Antecedent(universe_ToF_sensor_range, 'ToF_sensor_range')

# Asignar las funciones de pertenencia a los valores lingüísticos de ToF_sensor_range
ToF_sensor_range['CLOSE'] = fuzz.trapmf(universe_ToF_sensor_range, [0.0, 0.0, 0.2, 0.3])        # Close
ToF_sensor_range['NEAR'] = fuzz.trapmf(universe_ToF_sensor_range, [0.2, 0.3, 0.4, 0.5])         # Near
ToF_sensor_range['MED'] = fuzz.trapmf(universe_ToF_sensor_range, [0.4, 0.5, 0.6, 0.7])          # Medium
ToF_sensor_range['FAR'] = fuzz.trapmf(universe_ToF_sensor_range, [0.6, 0.7, 1.3, 1.4])          # Far
ToF_sensor_range['VFAR'] = fuzz.trapmf(universe_ToF_sensor_range, [1.3, 1.4, 10.0, 10.0])       # Very Far
```

```python
plt.figure(figsize=(13, 6))

# Graficar los conjuntos difusos
for label in ToF_sensor_range.terms:
    plt.plot(universe_ToF_sensor_range, ToF_sensor_range[label].mf, label=label)

# Etiquetas y título
plt.title('ToF sensor range')
plt.xlabel('Universo')
plt.ylabel('Grado de Pertenencia')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

#### **Sets for steering** $[-150, \dots, 150]$

```python
# Definir el universo para la variable difusa steering
universe_steering = np.arange(-150, 150, 0.1)
steering = ctrl.Antecedent(universe_steering, 'steering')

# Asignar las funciones de pertenencia a los valores lingüísticos de steering
steering['TTR'] = fuzz.trapmf(universe_steering, [-150.0, -150.0, -100.0, -90.0])       # Tight Right
steering['TR'] = fuzz.trapmf(universe_steering, [-100.0, -90.0, -60.0, -50.0])          # Right
steering['TSR'] = fuzz.trapmf(universe_steering, [-60.0, -50.0, -20.0, -10.0])          # Small Right
steering['TC'] = fuzz.trapmf(universe_steering, [-20.0, -10.0, 10.0, 20.0])             # Center
steering['TSL'] = fuzz.trapmf(universe_steering, [10.0, 20.0, 50.0, 60.0])              # Small Left
steering['TL'] = fuzz.trapmf(universe_steering, [50.0, 60.0, 90.0, 100.0])              # Left
steering['TTL'] = fuzz.trapmf(universe_steering, [90.0, 100.0, 150.0, 150.0])           # Tight Left
```

```python
plt.figure(figsize=(13, 6))

# Graficar los conjuntos difusos
for label in steering.terms:
    plt.plot(universe_steering, steering[label].mf, label=label)

# Etiquetas y título
plt.title('steering')
plt.xlabel('Universo')
plt.ylabel('Grado de Pertenencia')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

#### **Sets for goal differences** $[-\pi, \dots, \pi]$ (radians)

```python
plt.figure(figsize=(13, 6))

# Graficar los conjuntos difusos
for label in goal_differences.terms:
    plt.plot(universe_goal_differences, goal_differences[label].mf, label=label)

# Etiquetas y título
plt.title('goal differences')
plt.xlabel('Universo')
plt.ylabel('Grado de Pertenencia')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
